```python
from pathlib import Path

root = Path("D:/Data/mab/BGdataset/Paradigm_8020/lesion_mPFC_post")

# rename_map = {
#     ".pt": "_RNNfit_N32_LR0.001.pt",
#     "N32_LR0.pt": "_RNNfit_N32_LR0.pt",
#     "N48_LR0.pt": "_RNNfit_N48_LR0.pt",
# }

for folder in root.iterdir():
    if not folder.is_dir():
        continue

    base = folder.name  # e.g. BGF1
    # base = base.lower() + "_post"  # e.g. bgf1

    for pt_file in folder.glob("*.pt"):
        name = pt_file.name

        if name == f"{base}.pt":
            new_name = f"{base}_RNNfit_N32_LR0.001_E500_Swindow.pt"

        elif name == f"{base}N32_LR0.pt":
            new_name = f"{base}_RNNfit_N32_LR0.001_E500_Ssession.pt"

        elif name == f"{base}N48_LR0.pt":
            new_name = f"{base}_RNNfit_N48_LR0.001_E500_Ssession.pt"

        else:
            print(f"Skipping {pt_file}")
            continue

        new_path = folder / new_name
        print(f"{pt_file.name} -> {new_name}")
        pt_file.rename(new_path)
```
